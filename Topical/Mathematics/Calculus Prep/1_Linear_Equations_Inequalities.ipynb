1.1 Solve the equation: $3x-(x-7)=4x-5$

```python
# define the variable
x_1 = sp.symbols('x_1')

# define the equation
equation_1 = sp.Eq(3*x_1 - (x_1-7), 4*x_1 - 5)

# solve the equation
solution_1 = sp.solve(equation_1, x_1)

# output the solution
solution_1
```

1.2 Calculate the slope, $m$, of the line $4x-3y-9$.

```python
# define the variables
x_2, y_2 = sp.symbols('x_2 y_2')

# define the equation of the line
equation_2 = sp.Eq(4*x_2 - 3*y_2 - 9, 0)

# solve the equation for y (rearrange into y = mx + b form)
slope_intercept_form_2 = sp.solve(equation_2, y_2)

# extract the slope (coefficient of x)
slope_2 = slope_intercept_form_2[0].coeff(x_2)

# output the slope
slope_2
```

1.5 Write the equation of the line passing through the points $(-3,-8)$ and $(-6,2)$ in slope-intercept form.

```python
from fractions import Fraction

# Define the points
x1, y1 = -3, -8
x2, y2 = -6, 2

# Calculate the slope as a fraction
m = Fraction(y2 - y1, x2 - x1)

# Calculate the y-intercept (b) as a fraction
b = Fraction(y1) - m * Fraction(x1)

# Print the equation in slope-intercept form
print(f"The equation of the line is: y = {m}x + {b}")
```

1.6 Calculate the x- and y-intercepts of $3x-4y=-6$ and use them to graph the line.

```python
import numpy as np
import matplotlib.pyplot as plt

# Coefficients of the equation 3x - 4y = -6
a, b, c = 3, -4, -6

# Calculate x-intercept (y=0)
x_intercept = c / a  # (-6 / 3) = -2

# Calculate y-intercept (x=0)
y_intercept = c / b  # (-6 / -4) = 3/2

# Create the plot
x_values = np.linspace(-4, 4, 100)  # Range of x values for the line
y_values = (a * x_values + -c) / -b  # Rearrange to solve for y: y = (3x + 6)/4

# Plot the line
plt.plot(x_values, y_values, label="3x - 4y = -6")

# Plot the intercepts
plt.scatter(x_intercept, 0, color='red', zorder=5, label=f"x-intercept (-2, 0)")
plt.scatter(0, y_intercept, color='blue', zorder=5, label=f"y-intercept (0, 3/2)")

# Labels and title
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("Graph of the line 3x - 4y = -6")
plt.xlabel("x")
plt.ylabel("y")

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

1.7 Assume that line $p$ contains the point $(-3,1)$ afnd is parallel to $x-4y=1$.  Writethe equation of $p$ in slope-intercept form.

```python
from fractions import Fraction

# Slope of the line x - 4y = 1 is 1/4
m = Fraction(1, 4)

# Point (-3, 1) on the line p
x1, y1 = -3, 1

# Use point-slope form to calculate the y-intercept
b = y1 - m * x1

# Print the equation of the line in slope-intercept form
print(f"The equation of line p is: y = {m}x + {b}")
```

Bisecting Lines: Two lines bisect each other if they intersect at their midpoints, meaning each line divides the other into two equal parts at the point of intersection.

1.8 According to a basic Euclidean geometry theorem, the diagonals of a parallelogram bisect each other.  Demonstrate this theorem for parallelogram $ABCD$, where $A=(2,1)$, $B=(7,1)$, $C=(11,4)$, and $D=(6,4)$.

```python
# Function to calculate the midpoint of a line segment
def midpoint(x1, y1, x2, y2):
    return ((x1 + x2) / 2, (y1 + y2) / 2)

# Coordinates of the points A, B, C, and D
A = (2, 1)
B = (7, 1)
C = (11, 4)
D = (6, 4)

# Calculate the midpoints of diagonals AC and BD
midpoint_AC = midpoint(A[0], A[1], C[0], C[1])
midpoint_BD = midpoint(B[0], B[1], D[0], D[1])

# Display the results
print(f"Midpoint of diagonal AC: {midpoint_AC}")
print(f"Midpoint of diagonal BD: {midpoint_BD}")

# Check if the midpoints are the same
if midpoint_AC == midpoint_BD:
    print("The diagonals of the parallelogram bisect each other.")
else:
    print("The diagonals of the parallelogram do not bisect each other.")
```

1.9 Prove that $ABCD$ is a rhombus by verifying that its sides are congruent.

```python
import math

# Function to calculate the distance between two points
def distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Coordinates of the points A, B, C, and D
A = (2, 1)
B = (7, 1)
C = (11, 4)
D = (6, 4)

# Calculate the lengths of sides AB, BC, CD, and DA
AB = distance(A[0], A[1], B[0], B[1])
BC = distance(B[0], B[1], C[0], C[1])
CD = distance(C[0], C[1], D[0], D[1])
DA = distance(D[0], D[1], A[0], A[1])

# Display the side lengths
print(f"Length of AB: {AB}")
print(f"Length of BC: {BC}")
print(f"Length of CD: {CD}")
print(f"Length of DA: {DA}")

# Check if all sides are congruent
if AB == BC == CD == DA:
    print("ABCD is a rhombus because all sides are congruent.")
else:
    print("ABCD is not a rhombus because the sides are not congruent.")
```

1.10 Prove that $ABCD$ is a rhombus by verifying that its diagonals are perpendicular to one another.

```python
# Function to calculate the slope between two points
def slope(x1, y1, x2, y2):
    if x2 - x1 != 0:
        return (y2 - y1) / (x2 - x1)
    else:
        return None  # undefined slope (vertical line)

# Coordinates of the points A, B, C, and D
A = (2, 1)
B = (7, 1)
C = (11, 4)
D = (6, 4)

# Calculate the slopes of diagonals AC and BD
slope_AC = slope(A[0], A[1], C[0], C[1])
slope_BD = slope(B[0], B[1], D[0], D[1])

# Display the slopes
print(f"Slope of diagonal AC: {slope_AC}")
print(f"Slope of diagonal BD: {slope_BD}")

# Check if the product of the slopes is -1 (i.e., if the diagonals are perpendicular)
if slope_AC is not None and slope_BD is not None:
    if slope_AC * slope_BD == -1:
        print("The diagonals are perpendicular, so ABCD is a rhombus.")
    else:
        print("The diagonals are not perpendicular, so ABCD is not a rhombus.")
else:
    print("One of the diagonals is vertical, check for undefined slope.")
```

1.16 Graph the inequality: $-2\le{x}<3$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -3 to 4
x = np.linspace(-3, 4, 400)

# Create an array for y values (set y=0 for a number line)
y = np.zeros_like(x)

# Plot the number line
plt.plot(x, y, 'k-', lw=2)  # A horizontal line representing the number line

# Highlight the solution part of the number line (-2 <= x < 3)
x_solution = np.linspace(-2, 3, 300)
y_solution = np.zeros_like(x_solution)
plt.plot(x_solution, y_solution, 'b', lw=4, label=r"$-2 \leq x < 3$")  # Blue line for the solution

# Mark -2 with a closed circle
plt.plot(-2, 0, 'bo', markersize=10, label="Included (-2)")

# Mark 3 with an open circle
plt.plot(3, 0, 'bo', markersize=10, markerfacecolor='white', label="Excluded (3)")

# Set limits for x and hide y-axis
plt.xlim(-3, 4)
plt.ylim(-0.5, 0.5)

# Add grid for clarity
plt.grid(True)

# Remove y-axis labels and ticks
plt.gca().axes.get_yaxis().set_visible(False)

# Add x-axis labels
plt.xticks(np.arange(-3, 5, 1))

# Add title and legend
plt.title(r"Graph of the inequality: $-2 \leq x < 3$")
plt.legend()

# Show the plot
plt.show()
```

1.18 Solve and graph the inequality: $-7\le{1-2x}<11$.

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Solve the inequality using sympy
# Define the variable
x = sp.Symbol('x')

# Solve the first part of the inequality: -7 <= 1 - 2x
inequality1 = -7 <= 1 - 2*x
solution1 = sp.solve_univariate_inequality(inequality1, x, relational=False)

# Solve the second part of the inequality: 1 - 2x < 11
inequality2 = 1 - 2*x < 11
solution2 = sp.solve_univariate_inequality(inequality2, x, relational=False)

# Combine the solutions
solution = solution1.intersect(solution2)

# Display the solution
print(f"The solution to the inequality is: {solution}")

# Step 2: Graph the solution using matplotlib
# Create an array of x values from -6 to 5 (for the number line)
x_vals = np.linspace(-6, 5, 400)

# Create an array for y values (set y=0 for a number line)
y_vals = np.zeros_like(x_vals)

# Plot the number line
plt.plot(x_vals, y_vals, 'k-', lw=2)  # A horizontal line representing the number line

# Highlight the solution part of the number line (-5 < x <= 4)
x_solution = np.linspace(-5, 4, 300)
y_solution = np.zeros_like(x_solution)
plt.plot(x_solution, y_solution, 'b', lw=4, label=r"$-5 < x \leq 4$")  # Blue line for the solution

# Mark -5 with an open circle (since it's excluded from the solution)
plt.plot(-5, 0, 'bo', markersize=10, markerfacecolor='white', label="Excluded (-5)")

# Mark 4 with a closed circle (since it's included in the solution)
plt.plot(4, 0, 'bo', markersize=10, label="Included (4)")

# Set limits for x and hide y-axis
plt.xlim(-6, 5)
plt.ylim(-0.5, 0.5)

# Add grid for clarity
plt.grid(True)

# Remove y-axis labels and ticks
plt.gca().axes.get_yaxis().set_visible(False)

# Add x-axis labels
plt.xticks(np.arange(-6, 6, 1))

# Add title and legend
plt.title(r"Graph of the inequality: $-7 \leq 1 - 2x < 11$")
plt.legend()

# Show the plot
plt.show()
```

1.19 Graph the inequality: $y<-\frac{1}{3}x+2$
