```python
import matplotlib.pyplot as plt

# Original data for supply and demand
prices = [50, 60, 70, 80]  # Prices in cents
demand = [5, 4, 3, 2]      # Quantity demanded in millions
supply = [3, 4, 5, 6]      # Quantity supplied in millions

# Updated demand curve after $0.20 tax (20 cents)
taxed_demand_prices = [price - 20 for price in prices]  # Prices reduced by 20 cents

# Plot the original supply curve
plt.plot(supply, prices, label='Original Supply Curve', marker='o', linestyle='-', color='blue')

# Plot the original demand curve
plt.plot(demand, prices, label='Original Demand Curve', marker='o', linestyle='-', color='red')

# Plot the updated demand curve after tax
plt.plot(demand, taxed_demand_prices, label='Demand Curve After $0.20 Tax', marker='o', linestyle='--', color='green')

# Add labels and title
plt.title('Brownies Market: Supply and Demand Curves', fontsize=14)
plt.xlabel('Quantity (millions of brownies)', fontsize=12)
plt.ylabel('Price (cents per brownie)', fontsize=12)

# Add grid and legend
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend(fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
