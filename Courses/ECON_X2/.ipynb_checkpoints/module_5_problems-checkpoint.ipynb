```python
import matplotlib.pyplot as plt

# Given data
output = [30, 70, 120, 160, 190, 210, 220]
tfc = [1000] * len(output)  # Total fixed cost remains constant
tvc = [500, 1000, 1500, 2000, 2500, 3000, 3500]  # Total variable cost
tc = [tfc[i] + tvc[i] for i in range(len(output))]  # Total cost

# Plotting the cost curves
plt.figure(figsize=(8, 5))
plt.plot(output, tfc, label="Total Fixed Cost (TFC)", linestyle="dashed", color="red")
plt.plot(output, tvc, label="Total Variable Cost (TVC)", linestyle="dotted", color="blue")
plt.plot(output, tc, label="Total Cost (TC)", linestyle="solid", color="green")

# Labels and legend
plt.xlabel("Output (Surfboards)")
plt.ylabel("Cost ($)")
plt.title("Cost Curves for Sue's Surfboards")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Given data
output = [30, 70, 120, 160, 190, 210, 220]  # Output levels
tfc = 1000  # Total fixed cost (constant)
tvc = [500, 1000, 1500, 2000, 2500, 3000, 3500]  # Total variable cost
tc = [tfc + tvc[i] for i in range(len(output))]  # Total cost

# Compute AFC, AVC, ATC, MC
afc = [tfc / q for q in output]  # Average Fixed Cost
avc = [tvc[i] / output[i] for i in range(len(output))]  # Average Variable Cost
atc = [tc[i] / output[i] for i in range(len(output))]  # Average Total Cost

# Compute MC using TC / Q (set first MC as None since no prior data)
mc = [None]  # First MC undefined
for i in range(1, len(output)):
    delta_tc = tc[i] - tc[i-1]
    delta_q = output[i] - output[i-1]
    mc.append(delta_tc / delta_q)

# Plotting cost curves
plt.figure(figsize=(8, 5))
plt.plot(output, afc, label="Average Fixed Cost (AFC)", linestyle="dashed", color="red")
plt.plot(output, avc, label="Average Variable Cost (AVC)", linestyle="dotted", color="blue")
plt.plot(output, atc, label="Average Total Cost (ATC)", linestyle="solid", color="green")
plt.plot(output[1:], mc[1:], label="Marginal Cost (MC)", linestyle="dashdot", color="purple", marker="o")

# Labels and legend
plt.xlabel("Output (Surfboards)")
plt.ylabel("Cost ($)")
plt.title("Cost Per Unit Analysis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Given data
labor = [1, 2, 3, 4, 5, 6, 7]  # Workers per week
output = [30, 70, 120, 160, 190, 210, 220]  # Output per week
tvc = [500, 1000, 1500, 2000, 2500, 3000, 3500]  # Total variable cost

# Calculate AP and MP
ap = [output[i] / labor[i] for i in range(len(labor))]  # AP = Q / L
mp = [None]  # First MP undefined
for i in range(1, len(labor)):
    mp.append(output[i] - output[i-1])  # MP = Q / L

# Calculate AVC and MC
avc = [tvc[i] / output[i] for i in range(len(output))]  # AVC = TVC / Q
mc = [None]  # First MC undefined
for i in range(1, len(output)):
    delta_tc = tvc[i] - tvc[i-1]
    delta_q = output[i] - output[i-1]
    mc.append(delta_tc / delta_q)  # MC = TC / Q

# Plot AP & MP
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(labor, ap, label="Average Product (AP)", linestyle="dashed", color="blue", marker="o")
plt.plot(labor[1:], mp[1:], label="Marginal Product (MP)", linestyle="solid", color="green", marker="s")
plt.xlabel("Labor (Workers)")
plt.ylabel("Productivity (Output per Worker)")
plt.title("Average & Marginal Product")
plt.legend()
plt.grid(True)

# Plot AVC & MC
plt.subplot(1, 2, 2)
plt.plot(output, avc, label="Average Variable Cost (AVC)", linestyle="dashed", color="red", marker="o")
plt.plot(output[1:], mc[1:], label="Marginal Cost (MC)", linestyle="solid", color="purple", marker="s")
plt.xlabel("Output (Surfboards)")
plt.ylabel("Cost ($ per Surfboard)")
plt.title("Average & Marginal Cost")
plt.legend()
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```
