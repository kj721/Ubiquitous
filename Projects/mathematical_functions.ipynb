# Linear Functions

Form: $f(x)=mx+b$
$m$ is the slope (rate of change)
$b$ is the intercept (starting value)

It graphs as a straight line, and the change in $y$ is proportional to the change in $x$.


1. Finance: Interest earned on a simple savings account.
$I(t)=50t+1000$ where $t=\text{years}$.
2. Strategy (LIC immunization): Delivery cost per child with no scaling advantage.
$\text{Cost}(n)=2n+5000$ where $n=\text{children reached}$.

## Example 1: Finance Simple Interest
Formula: $I(t)=50t+1000$
$t$: time in years
50: interest per year
1000: intial principal

```python
def linear_fn(x, m, b):
    return m * x + b
```

```python
def plot_linear_functions(interest_rate=50, initial_principal=1000, cost_per_child=2, fixed_cost=5000):
    x_interest = np.linspace(0, 20, 200)
    x_children = np.linspace(0, 10000, 200)

    y_interest = linear_fn(x_interest, interest_rate, initial_principal)
    y_cost = linear_fn(x_children, cost_per_child, fixed_cost)

    fig, axs = plt.subplots(1, 2, figsize(15, 5))

    axs[0].plot(x_interest, y_interest, color='green', label='Interest')
    axs[0].set_title('Simple Interest Growth')
    axs[0].set_xlabel('Years')
    axs[0].set_ylabel('Total Balance (USD)')
    axs[0].grid(True)
    axs[0].legend()

    axs[1].plot(x_children, y_cost, color='blue', label='Delivery Cost')
    axs[1].set_title('Immunization Delivery Cost (Linear)')
    axs[1].set_xlabel('Children Reached')
    axs[1].set_ylabel('Total Cost (USD)')
    axs[1].grid(True)
    axs[1].legend()

    plt.tight_layout()
    plt.show()

interact(
    plot_linear_functions,
    interest_rate=FloatSlider(value=50, min=0, max=200, step=5, description='Interest/Year'),
    initial_principal=FloatSlider(value=1000, min=0, max=5000, step=100, description='Initial Principal'),
    cost_per_child=FloatSlider(value=2, min=0, max=10, step=0.1, description='Cost/Child'),
    fixed_cost=FloatSlider(value=5000, min=0, max=20000, step=500, description='Fixed Cost')
);
```

# Affine Functions

An affine function is a special type of linear function

$f(x)=ax+b$

Just like a linear function, but the key distinction is that:
- It doesn't necessarily go through the origin (i.e., when $x=0$, $f(x)=b$)
- It's often used to model situations where there is both a fixed component (b) and a variable component (a$\cdot$x)

## Example 1: Household Spending

You're modeling a household's monthly expenses:

$E(x)=1200+30x$

Where:
- $x$ is number of family members
- 1200 is base cost (housing, utilities)
- 30 is the marginal spending per person

## Example 2: Political Science/Strategy: Campaign Impact

Estimating the total number of voters reached by campaign work:

$V(h)=20000+400h$

Where:
- $h$ is hours of on-the-ground campaigning
- 20000 is the baseline awareness
- 400 is new voters reached per hour of campaigning

```python
def household_expenses(x, base_cost=1200, cost_per_person=30):
    return base_cost + cost_per_person * x

def voter_reach(hours, base_support=20000, voters_per_hour=400):
    return base_support + voters_per_hour * hours


```
