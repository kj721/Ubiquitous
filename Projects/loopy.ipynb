# Using `for` loops with charts

### Plotting multiple lines or datasets

```python
datasets = [
    [1, 2, 3, 4, 5],
    [2, 3, 5, 7, 11],
    [1, 4, 9, 16, 25]   
]
```

```python
for i, data in enumerate(datasets):
    plt.plot(data, label=f'Dataset {i+1}')
    
plt.title('Multiple Datasets')
plt.legend()
plt.show()
    
```

### Customizing multiple subplots

```python
# creating a figure with 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# sample data
data = [range(5), range(1, 6), range(2, 7)]

# use a loop for each subplot
for i, ax in enumerate(axs):
    ax.plot(data[i])
    ax.set_title(f'Subplot {i+1}')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    
plt.tight_layout()
plt.show()
```

### Managing labels, colors, and other properties

```python
data = [5, 7, 3]
labels = ['A', 'B', 'C']
colors = ['red', 'green', 'blue']

for i in range(len(data)):
    plt.bar(i, data[i], color=colors[i], label=labels[i])
    
plt.xticks(range(len(labels)), labels)
plt.legend()
plt.show()
```

```python
# Write a for loop that computes the factorial of a given number `n` 

def fact(number):
    if number < 0:
        return "The factorial of a negative number is undefined."
    total = 1
    for n in range(1, number+1):
        total *= n
    return total
    
print(fact(26))
```

```python
# write a loop that counts the number of vowels in a given string

def v_count(word):
    vowels = ['a','e','i','o','u']
    total = 0
    for char in word.lower():
        total += char in vowels
    return total

print(v_count("ZIVallltuoeslin"))
```

Step 3: Checking for Primality
Problem:
To determine if a number `i` is prime, we need to check if it has any divisors other than 1 and itself.

Brute-Force Approach (Not Optimal):
One way to check if a number `i` is prime is to try dividing it by every number from 2 up to `i` - 1. If i is divisible by any of these numbers, it is not prime.

However, this approach is inefficient, especially for larger numbers, because you are performing a lot of unnecessary checks.

Optimization: Checking Up to the Square Root of `i`
We can make this process much faster by only checking divisors up to the square root of `i`. Here's why:

Basic Division Property:

If `i` is divisible by some number `j`, then `i` = `j` * `k` for some integer `k`.
If both `j` and `k` were greater than the square root of `i`, then their product `j` * `k` would be greater than `i`, which is a contradiction. Therefore, at least one of these factors must be less than or equal to the square root of `i`.
Checking up to sqrt(`i`):

If we find that `i` is divisible by any number `j` between 2 and sqrt(`i`), then `i` is not prime.
If no divisors are found in this range, then `i` is prime.

Implementation:

In the code:
`for j in range(2, int(i ** 0.5) + 1):
    if i % j == 0:
        is_prime = False
        break
range(2, int(i ** 0.5) + 1):`

We loop through all numbers `j` from 2 up to `int(i ** 0.5) + 1`.
`int(i ** 0.5)` calculates the integer value of the square root of `i`.
We add 1 to include the square root itself in the range.
`if i % j == 0:`:

This checks if `i` is divisible by `j` (i.e., the remainder when `i` is divided by `j` is 0).
If `i` is divisible by any `j` in this range, `i` is not prime, so we set `is_prime` to `False` and break out of the loop (theres no need to check further).

Efficiency:

This method drastically reduces the number of checks we need to perform, making the algorithm much faster, especially for larger numbers.
Example:
Lets say we want to check if 29 is a prime number:

- The square root of 29 is approximately 5.39.
- We only need to check divisibility by 2, 3, 4, and 5.
- If 29 isnt divisible by any of these numbers, it must be prime (which it is).

Why This Works:
If a number greater than 1 is not prime, it can be factored into two smaller numbers. If neither factor is greater than the square root of the number, both must be less than or equal to the square root, hence the logic of checking only up to the square root.

Summary:
Efficiency: By checking divisibility only up to the square root of the number, we reduce the number of checks needed to determine if the number is prime.

Correctness: This approach ensures that we accurately identify prime numbers while avoiding unnecessary calculations.
