```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return 3 * x + 1

# x values for plotting the function
x = np.linspace(-1, 5, 400)
y = f(x)

# x values for the shaded area
x_fill = np.linspace(0, 4, 100)
y_fill = f(x_fill)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = 3x + 1', color='blue')
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.6, label='Area under curve (0 to 4)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Add labels and title
plt.title('Graph of y = 3x + 1 and Area Under Curve from x = 0 to x = 4')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.xlim(-1, 5)
plt.ylim(0, f(4) + 2)

# Save the graph as a JPG image
plt.savefig('graph_y_equals_3x_plus_1.jpg', format='jpg', dpi=300)

plt.show()
```
