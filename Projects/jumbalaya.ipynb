```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import random
```

### Normal Distribution

```python
# Assumptions
pmt = 400000
r = 0.1013
s = 0.0637
n = 20
sim = 1000000
```

```python
# Data generation and summary statistics
np.random.seed(42)
data_norm = np.random.normal(r, s, (sim, n))
calc_norm = pmt * (1 + data_norm)
mean_norm = np.mean(calc_norm)
max_norm = np.max(calc_norm)
min_norm = np.min(calc_norm)
print(f"The average simulated investment value at the end of {n} years is ${mean_norm:,.2f}")
print(f"The maximum simulated investment value at the end of {n} years is ${max_norm:,.2f}")
print(f"The minimum simulate investment value at the end of {n} years is ${min_norm:,.2f}")
```

```python
# Visualize the distribution
# Sort the final values to compute the CDF
sorted_values = np.sort(calc_norm)
cdf = np.arange(1, len(sorted_values) + 1) / len(sorted_values)

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(sorted_values, cdf, label="CDF", linewidth=2)
plt.axvline(x=np.median(calc_norm), color='red', linestyle='--', label="Median")
plt.axvline(x=np.percentile(calc_norm, 5), color='blue', linestyle='--', label="5th Percentile")
plt.axvline(x=np.percentile(calc_norm, 95), color='green', linestyle='--', label="95th Percentile")

# Customize plot aesthetics
plt.title("Cumulative Distribution Function of Final Portfolio Values", fontsize=16)
plt.xlabel("Final Portfolio Value ($)", fontsize=14)
plt.ylabel("Cumulative Probability", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Probabalization 
# Define the threshold
threshold = 1052

# Calculate the probability
probability = np.mean(calc_norm >= threshold)
print(f"The probability of the portfolio value being  ${threshold:,.2f} is {probability:.2%}.")
```
