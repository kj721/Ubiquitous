```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_investment = 650000  # starting amount
years = 15
expected_return = 0.07      # 7% expected return
std_dev = 0.10              # 10% standard deviation
num_simulations = 10000

# Monte Carlo simulation
np.random.seed(42)  # for reproducibility
final_values = []

for _ in range(num_simulations):
    yearly_returns = np.random.normal(expected_return, std_dev, years)
    future_value = initial_investment
    for r in yearly_returns:
        future_value *= (1 + r)
    final_values.append(future_value)

# Convert results to numpy array
final_values = np.array(final_values)

# Plotting the results
plt.hist(final_values, bins=100, edgecolor='black')
plt.title('Monte Carlo Simulation of Investment Value After 15 Years')
plt.xlabel('Future Value ($)')
plt.ylabel('Frequency')
plt.axvline(np.percentile(final_values, 5), color='r', linestyle='dashed', label='5th percentile')
plt.axvline(np.percentile(final_values, 95), color='g', linestyle='dashed', label='95th percentile')
plt.legend()
plt.grid(True)
plt.show()

# Basic stats
print(f"Mean Future Value: ${final_values.mean():,.2f}")
print(f"Median Future Value: ${np.median(final_values):,.2f}")
print(f"5th Percentile: ${np.percentile(final_values, 5):,.2f}")
print(f"95th Percentile: ${np.percentile(final_values, 95):,.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_investment = 650000
years = 15
expected_return = 0.07
std_dev = 0.10
num_simulations = 10000

# Vectorized Monte Carlo simulation
np.random.seed(42)
# Shape: (num_simulations, years)
random_returns = np.random.normal(expected_return, std_dev, (num_simulations, years))

# Compound the returns: (1 + r1) * (1 + r2) * ... * (1 + r15)
cumulative_growth = np.prod(1 + random_returns, axis=1)

# Final value = initial investment * compounded growth
final_values = initial_investment * cumulative_growth

# Plotting
plt.hist(final_values, bins=100, edgecolor='black')
plt.title('Monte Carlo Simulation of Investment Value After 15 Years')
plt.xlabel('Future Value ($)')
plt.ylabel('Frequency')
plt.axvline(np.percentile(final_values, 5), color='r', linestyle='dashed', label='5th percentile')
plt.axvline(np.percentile(final_values, 95), color='g', linestyle='dashed', label='95th percentile')
plt.legend()
plt.grid(True)
plt.show()

# Stats
print(f"Mean Future Value: ${final_values.mean():,.2f}")
print(f"Median Future Value: ${np.median(final_values):,.2f}")
print(f"5th Percentile: ${np.percentile(final_values, 5):,.2f}")
print(f"95th Percentile: ${np.percentile(final_values, 95):,.2f}")
```
