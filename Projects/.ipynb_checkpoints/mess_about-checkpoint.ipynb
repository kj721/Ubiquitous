```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import math
```

```python
st_bal = 500000
r = 0.07
std = 0.09
n = 75-49
sims = 1000
parameter = 600000
```

```python
prob = np.random.normal(r, std, sims)
data = st_bal * ((1 + prob)**n)
prob_results = np.sum(data > parameter)/len(data)
prob_results
```
