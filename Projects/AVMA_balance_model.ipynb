```python
import numpy as np
import matplotlib.pyplot as plt

# Initial parameters
initial_balance = 500e6  # $500 million
max_expenditure = 75e6  # $75 million
start_year = 2023
end_year = 2030
years = end_year - start_year

# Assumptions for simulation
num_simulations = 1000000
interest_rate_min = 0.03  # assuming a potential decrease in interest rates
interest_rate_max = 0.07

# Simulate scenarios
final_balances = []
for _ in range(num_simulations):
    balance = initial_balance
    for _ in range(years):
        # Simulate annual interest rate change and expenditure
        interest_rate = np.random.uniform(interest_rate_min, interest_rate_max)
        expenditure = np.random.uniform(0, max_expenditure / years)
        
        # Update balance
        balance = balance * (1 + (interest_rate / 12) - expenditure

        # Ensure expenditure does not exceed max_expenditure
        max_expenditure = max(0, max_expenditure - expenditure)

    final_balances.append(balance)

# Analyzing the results
mean_balance = np.mean(final_balances)
median_balance = np.median(final_balances)
percentile_5 = np.percentile(final_balances, 5)
percentile_95 = np.percentile(final_balances, 95)
mean_rate = np.mean(interest_rate)
max_balance = np.max(final_balances)
min_balance = np.min(final_balances)

# Plotting the distribution of final fund balances
plt.hist(final_balances, bins=50, color='blue', alpha=0.7)
plt.title("Distribution of Final Fund Balances by 2030")
plt.xlabel("Fund Balance ($ in Scientific Notation)")
plt.ylabel("Frequency")
plt.axvline(mean_balance, color='red', linestyle='dashed', linewidth=2, label='Mean Balance')
plt.axvline(median_balance, color='green', linestyle='dashed', linewidth=2, label='Median Balance')
plt.legend()
plt.show()

(mean_balance, median_balance, percentile_5, percentile_95)
```
